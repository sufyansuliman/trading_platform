```python
import pandas as pd 
spotMil = pd.read_csv("tracks_features.csv")
spot200 = pd.read_csv("Spot200.csv")
spotMil.head()
```

```python
spot200["id"] = spot200["Link"].str.replace(pat='^(.*[\\\/])', repl='', regex=True)
spot200.head()
```

```python
spotMil["top200"] = spotMil["id"].isin(spot200["id"]).astype(int)
spotMil.head()
```

```python
spotMil['year'] = spotMil['year'].astype(int)
spotMil = spotMil[spotMil['year'] >= 2017]
spotMil = spotMil[spotMil['year'] <= 2020]
```

```python
spotMilOnes = spotMil[spotMil['top200'] == 1]
spotMilZeros = spotMil[spotMil['top200'] == 0]
spotMilOnes.head()
```

```python
spotMilZeros.head()
```

```python
sampledZeros = spotMilZeros.sample(n = 2000)
sampledZeros.head()
```

```python
combinedDF = pd.concat([sampledZeros, spotMilOnes], axis = 0)
combinedDF.head()
```

```python
combinedDF.to_csv('SpotifySamples.csv')
```
